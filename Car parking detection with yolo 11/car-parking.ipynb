```python
image_path = "/kaggle/input/pklot-dataset/valid/2012-09-11_15_36_32_jpg.rf.b2b1767cafabff6378468d0d71c1ecd2.jpg"
results = model(image_path)

vacant_count = 0
occupied_count = 0

for box in results[0].boxes:
    class_name = results[0].names[int(box.cls)]
    if class_name == "Vacant":
        vacant_count += 1
    elif class_name == "Car":
        occupied_count += 1

total_spots = vacant_count + occupied_count

print(f"Total Spots: {total_spots}")
print(f"Vacant Spots: {vacant_count}")
print(f"Occupied Spots: {occupied_count}")

results[0].show()
results[0].save("output_image_with_counts.jpg")
```
